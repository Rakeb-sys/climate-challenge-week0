Outlier Detection & Basic Cleaning

```python
import pandas as pd
import numpy as np

cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']

# Compute Z-scores
z_scores = (df[cols] - df[cols].mean()) / df[cols].std()

# Flag outliers
outliers = (np.abs(z_scores) > 3)

# Count rows with any outlier
outlier_rows = outliers.any(axis=1)
outlier_count = outlier_rows.sum()

print(f"Number of rows with |Z| > 3: {outlier_count}")
```

```python
df_capped = df.copy()

for col in cols:
    mean = df[col].mean()
    std = df[col].std()
    upper = mean + 3 * std
    lower = mean - 3 * std
    
    df_capped[col] = df[col].clip(lower, upper)
```

```python
# Drop rows with >30% missing values
threshold = int(df_capped.shape[1] * 0.7)
df_clean = df_capped.dropna(thresh=threshold)

# Forward fill weather variables
df_clean[cols] = df_clean[cols].ffill()
```

```python
country = "tanzania"
df_clean.to_csv(f"../../dataset/{country}_clean.csv", index=False)
```

Time Series Analysis

```python
# Convert Date column to datetime
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
# Set as index
df_clean.set_index('Date', inplace=True)
```

```python
print(df_clean.index)
```

```python
# Create monthly aggregates
monthly = df_clean.resample('ME').mean(numeric_only=True)

# Monthly precipitation (sum, not mean)
monthly_precip = df_clean.resample('ME')['PRECTOTCORR'].sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(monthly.index, monthly['T2M'])

# Annotate extremes
max_idx = monthly['T2M'].idxmax()
min_idx = monthly['T2M'].idxmin()

plt.scatter(max_idx, monthly.loc[max_idx,'T2M'], color='red')
plt.scatter(min_idx, monthly.loc[min_idx,'T2M'], color='blue')

plt.title("Monthly Average T2M")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

Comment: This code performs exploratory data analysis (EDA) on the Sudan climate dataset.

Finally, it visualizes the monthly average temperature (T2M) and highlights the maximum and minimum points.

The Mpnthly average T2M plot shows the overall trend and seasonal patterns in temperature over time. 
The red point around 2022 indicates the month with the highest average temperature,

Correlation & Relationship Analysis

```python
import seaborn as sns
numeric_df = df_clean.select_dtypes(include='number')
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

Correlations:

    1. WS2M vs WS2M_MAX
        Relationship: Strong Positive

        Interpretation: 
        As the average daily wind speed (WS2M) increases, the maximum gust speed (WS2M_MAX) for that day increases.

    2. T2M  vs PS 
       Relationship: Strong Negative

        Interpretation: 
        AS T2M increases, PS also decreases, As temperature rises, surface pressure drops. 

    3. RH2M vs T2M_RANGE
       Relationship: Strong Negative

        Interpretation: 
        As days RH2M  increases, inversily T2M_RANGE decreases significantly. This is because moisture in the air prevents the ground from heating up too fast during the day or cooling down too fast at night. Conversely, your highest temperature swings happen on the driest days.
